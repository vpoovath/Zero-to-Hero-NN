## Intuitive Takeaway
What do we take away from this derivative exercise? 

When we 'nudge' a specific input by some very small amount $h$, we see how the value of the (scalar) function changes w.r.t. to that input and 'nudge'. 

And if we compute the 'slope' by using the approximation $\frac{d2-d1}{h}$, we can see what direction and by how much the function 'moved'.

# Value Object
Create a value object and build out the methods that can support mathematical operations.

```python
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        """
        data = 
        _children =
        _op =
        label = 
        """
        self.data = data
        self.grad = 0.0 # initialize gradeints to zero
        self._prev = set(_children)
        self._op = _op
        self.label = label
        self._backward = lambda:None
    
    def __repr__(self):
        """
        Print out nicer expr when inspecting Value obj
        """
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        out = Value(self.data + other.data, (self, other), "+")
        
        def _backward():
            self.grad  = 1.0 * out.grad
            other.grad = 1.0 * out.grad
        out._backward = _backward
        
        return out
    
    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), "*")
        
        def _backward():
            self.grad = other.data * out.grad
            other.grad = self.data * out.grad
        out._backward = _backward
        
        return out
    
    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x)+1)
        out = Value(t, (self,), "tanh")
        
        def _backward():
            self.grad = (1-t**2)*out.grad
        out._backward = _backward
        
        return out
    
    def backward(self):
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
                
        build_topo(self)
        
        self.grad = 1.0
        for node in reversed(topo):
            node._backward()


# # what happens here is that Python will call the following dunder func to carry out addition:
# # a.__add__(b)
# c = a + b
# print(c.data)

# # what happens here is that Python will call the following dunder func to carry out addition:
# # a.__mult__(b)
# d = a * b
# print(d.data)

a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(10.0, label='c')
e = a*b; e.label='e'
d = e + c; d.label='d'
f = Value(-2.0, label='f')
L = d*f; L.label='L'
```

```python
from graphviz import Digraph

def trace(root):
    """
    Builds a set of all nodes and edges in a graph
    """
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child,v))
                build(child)
    build(root)
    return nodes, edges
    
def draw_dot(root):
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR=Left to right
    
    nodes, edges = trace(root)
    for n in nodes:
        uid = str(id(n))
        #for any value in the graph, create a rectangular {'record'} node for it
        dot.node(name=uid, label="{ %s | data %.4f | grad %.4f}" % (n.label, n.data, n.grad), shape='record')
        if n._op:
            # if this value is a result of some operation, create an op node for it
            dot.node(name = uid + n._op, label = n._op)
            # and connect this node to it
            dot.edge(uid + n._op, uid)
        
    for n1, n2 in edges:
        # connect n1 to the op node of n2
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)
    
    return dot
```

## Manual Steps of Backpropagation
Want to look at the derivative of the loss function (L) w.r.t to its leaf nodes (weights in neural network). 

Note that we won't take derivative w.r.t. to inputs because inputs are fixed.

Here we are just manually setting the derivative of the loss function w.r.t. to leaf nodes using basic calculus derivations. The point of this exercise is to go locally to each node and apply the chain node. This will let you see how the loss function is affected by a change in a given node as well as an intuition on how gradient values are propagated through each node from the loss function.

Through this exericse, you will see that backpropagation is a recursive application of chain rule through a graph. It's one thing to watch a lecture on this, but do the math by hand to buid the muscle memory.

If you forget your derivatives, go back to first principles and derive the derivative. For example, taking dL/dd:

$\begin{align}
&\frac{f(x+h)-f(x)}{h}\\
&=\frac{L(x+h)-L(x)}{h}\\
&=\frac{(d+h)*f-d*f}{h}\\
&=\frac{df+hf-df}{h}\\
&=\frac{hf}{h}\\
&=f\\
\end{align}$


For example:

Note: $L = d * f$

Note: $L = d * f = (e+c) * f$

$\begin{align}
\frac{dL}{dL} & = 1\\
\frac{dL}{dd} & = f\\
\frac{dL}{df} & = d\\
\frac{dL}{dc} & = \frac{dL}{dd} * \frac{dd}{dc} \\
&= f * 1 \\
&= f\\
\frac{dL}{de} & = \frac{dL}{dd} * \frac{dd}{de} \\
&= f * 1 \\
&= f\\
\frac{dL}{da} & = \frac{dL}{de} * \frac{de}{da} \\
& = f*b\\
\frac{dL}{db} & = \frac{dL}{de} * \frac{de}{db} \\
& = f*a\\
\end{align}$

```python
def lol():
    """
    Little toy function to see what the derivative of L is w.r.t. to whatever
    variable you want. Just go in here and add 'h' to the respective variable.
    
    When you don't add 'h' to any one of the variables, derivative evals to 0.
    
    dL/dx = (L(x+h)-L(x))/h
    """
    h = 0.001
    
    # Evaluation of L(x)
    # where 'x' is a vector of all these vars
    a = Value(2.0, label='a')
    b = Value(-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b;
    d = e + c
    f = Value(-2.0, label='f')
    L1 = d*f
    
    # Just create labels for nodes that don't have one
    e.label='e'
    d.label='d'
    L1.label='L1'
    
    # Evaluation of L(x+h)
    a = Value(2.0, label='a')
    b = Value(-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b
    d = e + c
#     d.data = d.data + h # this helps evaluate dL/dd
    f = Value(-2.0, label='f')
#     f.data = f.data+h # this helps evaluate dL/df
    L = d*f
    L2 = Value(L.data, label='L2')
    
    e.label='e'
    d.label='d'
    L2.label='L2'
    
    gradL = (L2.data-L1.data)/h
    print(gradL)
    
lol()
```

## Single Optimization Step:
Now we nudge the values of the leaf nodes in the direction of their respective gradients. If we do this we expect the value of the loss function to increase (i.e. become less negative and move towards zero).

We define a uniform step size to determine how much of a nudge we want each of the leaf nodes to make.

After we make a nudge, we then manually compute the forward pass so that the loss function can be updated accordingly. 

This in essence is the training step of a neural net.

```python
step_size = 0.01
a.data += a.grad * step_size
b.data += b.grad * step_size
c.data += c.grad * step_size
# d.data += d.grad * step_size
# e.data += e.grad * step_size
f.data += f.grad * step_size

e = a * b
d = e + c
L = d * f

print(L)
```

```python
# inputs x1,x2
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')
# weights w1,w2
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')
# Bias of the neuron
b = Value(6.8813735870195432, label='b')
# intermediate nodes
# x1*w1 + x2*w2 + b
x1w1 = x1*w1; x1w1.label = 'x1*w1'
x2w2 = x2*w2; x2w2.label = 'x2*w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label = 'x1*w1 + x2*w2'
n = x1w1x2w2 + b; n.label = 'n'
o = n.tanh(); o.label = 'o' # tanh() implemented in Value obj
```
